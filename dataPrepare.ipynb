```python
import pandas as pd
import sqlite3
```

```python
house_data_to_load = "Resources/Suburb_HouseV2021A.csv"
house_data = pd.read_csv(house_data_to_load)
house_data.head(5)
```

```python
house_data = house_data.drop(index=([0,1,2,3])).copy()
house_data.head(5)
```

```python
house_data = house_data.rename(columns={'Suburb House  Price - 2021':'Suburb',
    'Unnamed: 1':'2011','Unnamed: 2':'2012','Unnamed: 3':'2013','Unnamed: 4':'2014','Unnamed: 5':'2015',
    'Unnamed: 6':'2016','Unnamed: 7':'2017','Unnamed: 8':'2018','Unnamed: 9':'2019','Unnamed: 10':'2020',
    'Unnamed: 11':'2021','Unnamed: 12':'2022','Unnamed: 13':'2020-2021','Unnamed: 14':'2011-2021',
    'Unnamed: 15':'Growth PA'} ).copy()
house_data.head(5)
```

```python
house_data = house_data.drop(columns=['Unnamed: 16']).copy()
house_data.head(5)
```

```python
import numpy as np
house_data1 = house_data.replace('-',np.NaN).copy()
#house_data = house_data.replace('NA',np.NaN).copy() 

#house_data.loc[house_data['Suburb'] =='BEEAC']
```

```python
for year in range(2011, 2022):
    column=f'{year}-{year+1}'

    house_data1[column]=round(house_data1.apply(lambda x:(float(x[f'{year+1}'])-float(x[f'{year}'])) / float(x[f'{year}']) * 100,axis=1),1)
```

```python
house_data1
```

```python
# Create Tables
conn = sqlite3.connect('static/data/house.sqlite')

c = conn.cursor()

c.execute('''DROP TABLE change;''')
c.execute('''DROP TABLE median_house;''')
""" c.execute('''CREATE TABLE year_period
       (year_period_id  CHAR(3) PRIMARY KEY     NOT NULL,
       NAME CHAR(50)    NOT NULL);''') """

c.execute('''CREATE TABLE change
       (suburb  CHAR(50) NOT NULL,
       period   char(50) NOT NULL,
       change_percent Numeric,
       primary key (suburb,period));''')

c.execute('''CREATE TABLE median_house
       (suburb  CHAR(50) NOT NULL,
       year integer NOT NULL,
       median Numeric,
       primary key (suburb,year));''')
print ("tables created")
```

```python
for ind, row in house_data1.iterrows():
    # Insert into table median_house
    for year in range(2011, 2023):
        print(row['Suburb'],year, row[f'{year}'])
        if ( pd.notnull(row[f'{year}'])):
            SQL = f"INSERT INTO median_house (suburb,year,median) VALUES ( '{row['Suburb']}', {f'{year}'}, {row[f'{year}']} )"
            print(SQL)
            c.execute(SQL)
    # Insert into table change

    for period in ['2011-2021','Growth PA']:
        if ( pd.notnull(row[period])):
            SQL = f"INSERT INTO change (suburb,period,change_percent) VALUES ( '{row['Suburb']}', '{period}', {row[period]} )"
            print(SQL)
            c.execute(SQL)

    for year in range(2011,2022):
        column=f'{year}-{year+1}'
        if (pd.notnull(row[column])):
            SQL = f"INSERT INTO change (suburb,period,change_percent) VALUES ( '{row['Suburb']}', '{column}', {row[column]} )"
            print(SQL)
            c.execute(SQL)

```
